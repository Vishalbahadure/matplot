Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide variety of plotting functions and features for generating high-quality graphs, charts, and plots for data analysis and visualization tasks.

Matplotlib is commonly used because it offers:

Ease of use: Matplotlib provides a simple interface for creating a wide range of plots with just a few lines of code.
Customizability: Users have fine-grained control over every aspect of their plots, allowing them to customize colors, fonts, labels, axes, and more.
Versatility: Matplotlib supports a wide range of plot types and can be used for various purposes, including exploratory data analysis, scientific plotting, and creating publication-quality figures.
Integration: It integrates seamlessly with other Python libraries like NumPy, pandas, and SciPy, making it a powerful tool for data visualization in conjunction with data manipulation and analysis.
Community and Support: Matplotlib has a large and active user community, which means that there is extensive documentation, tutorials, and resources available to help users get started and troubleshoot any issues they encounter.
Some common plots that can be created using the Pyplot module of Matplotlib include:

Line Plot: Used to visualize the relationship between two continuous variables by connecting data points with straight lines.
Scatter Plot: Displays individual data points as markers, allowing users to identify patterns, clusters, or relationships between two variables.
Histogram: Represents the distribution of a single variable by dividing it into bins and counting the number of observations in each bin.
Bar Plot: Displays categorical data with rectangular bars whose heights or lengths represent the frequency or magnitude of each category.
Pie Chart: Represents categorical data as a circular graph, with each category represented by a slice of the pie whose size is proportional to its frequency or proportion within the dataset.


```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot."""

A scatter plot is a type of plot used to visualize the relationship between two continuous variables.
It displays individual data points as markers on a two-dimensional plane, with one variable plotted along the x-axis and the other 
variable plotted along the y-axis. Scatter plots are useful for identifying patterns, trends, or correlations between variables.
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))


plt.figure(figsize=(8, 6))  
plt.scatter(x, y, color='blue', alpha=0.8)  


plt.title('Scatter Plot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')


plt.grid(True) \

```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""

The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure.
It allows you to arrange plots in a grid-like fashion, specifying the number of rows and columns, and the position of each subplot within the grid.
This is particularly useful when you want to compare multiple plots side by side or when you want to organize your visualizations neatly.
```

```python

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])


x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])


x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])


x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.figure(figsize=(10, 8)) 
plt.subplot(2, 2, 1) 
plt.plot(x1, y1, marker='o', color='blue')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2) 
plt.plot(x2, y2, marker='s', color='red')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)  
plt.plot(x3, y3, marker='^', color='green')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  
plt.plot(x4, y4, marker='d', color='orange')
plt.title('Line Plot 4')

plt.tight_layout()  
plt.show()
```

```python
"""Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])"""

A bar plot is a type of plot that presents categorical data with rectangular bars, where the length or height of each bar represents the magnitude or frequency of the corresponding category.
It is commonly used to compare the values of different categories or to visualize the distribution of a categorical variable.

Bar plots are useful for the following purposes:

Comparison: Bar plots allow easy comparison of values across different categories, making it straightforward to identify trends or differences.
Visualization of distributions: Bar plots can be used to visualize the distribution of categorical variables, showing the frequency or proportion of each category.
Presentation of data: Bar plots are visually appealing and easy to understand, making them suitable for presentations and reports.
```

```python

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.figure(figsize=(8, 6))  
plt.bar(company, profit, color='skyblue')  
plt.title('Bar Plot of Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 6))  
plt.barh(company, profit, color='lightgreen')
plt.title('Horizontal Bar Plot of Company Profits')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.grid(True)
plt.show()
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)"""

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset through five summary statistics: the minimum, 
first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a concise way to visualize the central tendency, spread, and variability of the data, as well as identify any outliers.

Box plots are useful for the following purposes:

Summary of distribution: Box plots summarize the distribution of data, including the median, quartiles, and range, providing insights into the central tendency and variability of the dataset.
Comparison of distributions: Box plots allow for easy comparison of distributions between multiple groups or datasets, enabling analysts to identify differences in central tendency, spread, and skewness.
Detection of outliers: Box plots provide a visual indication of potential outliers in the dataset, making it easier to identify observations that fall significantly outside the typical range.
```

```python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


data = [box1, box2]


plt.figure(figsize=(8, 6))  
plt.boxplot(data, patch_artist=True, notch=True, vert=False) 
plt.title('Box Plot of Box1 and Box2')
plt.xlabel('Value')
plt.ylabel('Box')
plt.yticks([1, 2], ['Box1', 'Box2'])  
plt.grid(True)  
plt.show()
```
